## 0) Background <a name='background'/>

The goal of this study is to provide a principled measure of "optimality" in human decision-making that explicitely takes into account the strategy of the decision maker. There is a wealth of research proposing that humans make irrational, suboptimal decisions. Examples such as framing effects, violations of transitivity, and conjunction fallacies are all prominent examples of cases in which people make decisions that deviate from the most accurate policy.

Our interest is primarily in the domain of inference, situations in which people use noisy observations to form "mental models" of latent processes (e.g., using past trends to predict the weather tomorrow, or to pick investment strategies). People often perform suboptimally in inference environments, such that the decisions they make are less accurate than ideal observers. For example, people often "probability match", tending to make predictions that match the actually probability that the events they are predicting will occur (e.g., if trying to predict the next flip from a coin that flips "heads" 80% of the time, people will predict "heads" 80% of the time; see [Vulkan, 2002](https://onlinelibrary.wiley.com/doi/abs/10.1111/1467-6419.00106)). In these environments, the most accurate strategy is to identify the most likely outcome (e.g., "heads" if a coin in 80% biased towards heads) and exclusively predict this outcome (e.g., exclusively guess "heads", resulting and average 80% correct predictions). Although some people do adopt these types of "maximizing" strategies, the majority adopt less accurate "matching" strategies (e.g., responding "heads" 80% of the time, resulting in 64% correct predictions). 

Thus, optimality is often measured in terms of accuracy, with strategies that are more accurate being considered more optimal. However, this definition omits an important dimension known to play a role in human decision-making: the cognitive effort required to implement the optimal strategy. Optimal solutions can often incure high computational costs, even if they provide marginal improvements in accuracy ([Tavoni et al., 2019](https://www.sciencedirect.com/science/article/pii/S235215461930066X); [Findling et al., 2020](https://www.nature.com/articles/s41593-019-0518-9)). As such, a proper definition of optimality should include not only accuracy, but the tradeoff between accuracy and the cost/effort required to acheive this accuracy.

This notion is at the core of the theory of "bounded rationality" ([Simon, 2000](https://link.springer.com/article/10.1007%252FBF02512227); [Gigerenzer & Selten, 2002](https://pure.mpg.de/rest/items/item_2102373/component/file_2102372/content)). Bounded rationality proposes that seemingly irrational or suboptimal behavior results from cognitive resource limitations, whereby people look for "good enough" type strategies that acheive acceptable levels of accuracy. While these strategies do not maximize acheivable accuracy, they may provide enough accuracy in exchange for lower resource costs, and therefore be optimal with respect to the agent's goal of maximizing accuracy while minimizing resource use. This idea has provided a strong foundation for shifting thoughts about what consitutes "optimality" in decision-making.

However, while the framework of "bounded rationality" offers an important consideration for measures of optimal behavior, it does not provide a formal framework to measure or assess trade-offs between effort/resources use and accuracy. Here we propose that the information bottleneck method provides such a framework ([Tishby et al., 2001](https://arxiv.org/abs/physics/0004057)). The information bottleneck method is an information theoretic analysis of trade-offs between the amount of information a model encodes (effort) and the model's predictive performance (accuracy). This method has successfully been applied to a number of different disciplines (e.g., [machine learning](https://ieeexplore.ieee.org/abstract/document/7133169?casa_token=ZeWKoYj5x1EAAAAA:kfKTT6gatuUGqZjfXrO-FAFkWFjg4MlTSCJKwKs6ltaNCCs80aQs8Stq7nCe7Es5dnNCRXt0), [biology](https://www.pnas.org/content/112/22/6908.short), etc.) to specifically measure performance trade-offs. As such, we propose using this method in the domain of human learning to assess how optimally humans use information from their environment to make predictions.

## 1) Hypotheses<a name='hypotheses'/>

The main goal of this project is to adapt and apply the information bottleneck method to human behavioral tasks, with the specific goal of identifying how optimally humans trade-off effort and accuracy during inference. Our specific implementation of the method is expanded on in detail in section 3. For all hypotheses, we will be assessing the relationship between the amount of information a person is encoding (their information complexity) and the accuracy of their strategy (predictive accuracy). We will then be comparing where they fall in this complexity/accuracy space to the true optimal information bottleneck bound. People that fall closer to/farther from this bound will be measured as making more/less optimal use of the information they are encoding.

We will use this method to test two primary hypotheses:

### 1.1) Probability matching-like behavior results from optimal tradeoff between complexity and accuracy

We hypothesize that "probability matching"-like strategies require less information compelxity than "maximizing" strategies. As a result, these matching strategies will not be as accurate as maximizing strategies, but we still be optimal, insofar as they make maximally effective use of the reduced information being encoded. We hypothesize that while people may show varying degrees of probability matching-like behavior, they will be similarly close to the maximum acheivable accuracy given the amount of past information they choose to encode. There are 3 specific hypotheses we will test:

* **1) People who use "maximizing"-like strategies have higher information complexity than those use "matching" strategies.** We predict that people who "maximize" will use strategies that encode more information about the past than people who probability match.

* **2) People using more information complex strategies are more accurate.** We predict that people who encode more information from the past will acheive higher levels of predictive accuracy.

* **3) People fall equally far from the optimal bound, regardless of their strategy/information complexity.** We predict that simpler "matching" strategies will make equally effective use of the relatively smaller information being encoded as more information complex "matching" strategies.

The task used to test these hypotheses is outlined in section 2.2.1.

### 1.2) Inference is more optimal in predictibly stable environments than in predictably unstable environments

We have previously found an asymmetry in human inference processes, such that humans tend to learn well in stable environments, when contigencies do not change often (low hazard rate environments), but not well in unstable environments, when contigencies change frequently but predictably (high hazard rate environments; [Filipowicz et al., 2020](https://elifesciences.org/articles/57872)). Pilot analyses on the data from Filipowicz et al., 2020 (shown in <a href=#addInfo>section 4</a>) suggest that this reduction in accuracy resuts from less optimal use of the available information in predictably unstable environments. Moreover, these pilot analyses propose that people using more complex beliefs in unstable environments favor strategies that use information from the immediately preceding trial, rather than from trials farther in the past, as would be expected from more optimal models. There are therefore 4 specific hypotheses we will test:

* **1) People use less complex beliefs in predictably unstable environments than in stable environments.** Based on pilot analyses on the data from Filipowicz et al., 2020, people tend to use less of the available information in predictably unstable environments than in stable environments.

* **2) People fall equally far from the optimal bound in predictably stable environments, no matter their information complexity.** Pilot analyses on data from Filipowicz et al., 2020 show that while accuracy varies across subjects in relatively stable environments, subjects make equally effective use of the information they encode, no matter how complex.

* **3) People fall farther from the optimal bound as they get more information complex in predictably unstable environments because they follow "one-back" strategies.** Pilot analyses on data from Filipowicz et al., 2020 propose that people make less effective use of the information they encode in predictably unstable environments because subjects encode information primarily about the last observed data-point, which is a suboptimal strategy. We therefore expect that subjects will become increasingly suboptimal as they use more complex strategies in predictably unstable environments.

* **4) People use information in predictably unstable environments more optimally when cued to consider relevant latent statistics.** We predict that some of the suboptimalities in the unstable environments stems from strategies that are not appropriate for the particular environment. We therefore expected that more explicit cues about the correct strategies in these environments improve how optimally people use the information they encode.

The tasks used to test these hypotheses are outlined in sections 2.2.2 and 2.2.3.

## 2) Methods<a name='methods'/>

### 2.1) Participants

Participants will be recruited on Prolific and we will recruit 70 participants in each of three studies (210 total). In the low hazard condition of Filipowicz et al., 2020, we observed no correlation between information complexity and distance from the bound, with a regression Bayes factor comparing evidence for the alternative vs the null equal to 0.13 (i.e., a moderate effect; correlation shown in section 4). Given that we predict a null effect between complexity and distance from the bound in certain circumstances, we used the properties of this pilot correlation to determine the number of subjects needed to detect moderate evidence for the null with >90% power. 

To conduct our power analysis, we fit the correlation between information complexity and distance from the bound using the Bayesian regression methods outlined in section 3.4. We then used these fits to randomly generate 100 simulated data sets of $n$ new data points, and re-fit these simulated data sets using the same Bayesian regression methods to identify how often our analysis provided at least moderate evidence for the null (Bayes factor <.3) for different samples sizes $n$ (see 3.4 for details on how Bayes factor was computed). Using these simulation methods we found that $n=70$ provides >90% power to identify moderate evidence for the null, and 50% power to identify strong evidence for the null.

Similar power analyses using $n=70$ for our directional effects (i.e., correlation between accuracy and complexity, and between complexity and distance from the bound in the high hazard environment) reveal >99% power to identify very strong evidence for a correlation (Bayes factor >> 100). Given that we are interested in both null and directional effects, using $n=70$ ensures that our sample size is sufficiently large to detect at least moderate evidence for both types of effects.

For recruitment, we will have the following inclusion criteria: participants on Prolific must have completed at least 20 tasks, and have a 95% task approval rating or greater. Of the participants that start the task, we will exclude any participants that do not complete the entire experiment, without looking at their data, and collect data from participants until our 70 participant per experiment criterion is met. 

There may additionally be participants that do not perform the task as instructed (e.g., press the same button throughout the whole task). These participants do not strictly cause a problem for our analysis method - participants not engaging with the task are measured as encoding no information about the task and acheiving no accuracy. Nevertheless, we will attempt to identify paticipants using these strategies by examining their response distributions. On average, participants should make an equal number of both responses available. In 20 pilot participants, 3 chose one of the options on more than 60% of trials. As such, to avoid including too many participants with biases towards either option, we will include up to 10 out of 70 participants per experiment making one response on greater than 60% of trials. Any additional participants meeting this criterion will be excluded and another participant will be recruited until our 70 participant target is reached.

### 2.2) Study design

Participants will perform one of three bead estimation tasks. In the tasks, participants will see a white or black bead drawn from one of two jars on each trial. Participants will be told the proportion of white/black beads in each jar, and informed how often the jars the beads will be pulled from will switch (i.e., the hazard rate). A participant's goal will be to use the history of beads they saw on previous trials, and information about jar switch frequency, to predict the color of the next bead as accurately as possible. Using this environment we will have three different experiments.

#### 2.2.1) Markov bead prediction task

This first experiment is meant to test the hypotheses outlined in section 1.1. It will have 70 participants predict the next bead in an environment where one jar contains 100% black beads and the other contains 100% white beads. Participants will perform two blocks of 500 trials, with different jar switch rates, or "hazard" rates, in each block. One block will be a "low hazard" condition, in which the jar switch probability will be .32. The other block will be a "high hazard" condition, in which the jar switch probability will be .68. At the start of each task block, participants will be explicitely told the hazard rate, and see two jars with 100% black/white beads throughout the experiment. The order of the high and low hazard rate conditions will be counterbalanced between subjects. Additionally, although the sequence of beads will be generated randomly for each subject, the exact position of the low probability events (jar switches in the low hazard condition and jar repeats in the high hazard condition) will be kept identical, thus preserving the predictability between the two environments and reducing performance differences due to random variations in the different sequences.

<img src="./figs/MarkovTask-15.png" alt="Drawing" style="width: 700px;"/>

These hazard rates match the overall mean number of trial-to-trial bead switches expected in the other task environments, which will have hazard rates of .01/.99 in the low and high hazard conditions respectively, and jar probabilities of 80% (identical to the high and low hazard conditions in Filipowicz et al., 2020). The average switch rate can be computed by accounting got the hazard rate and jar probabilities in each condition. Given a hazard rate $H$ and a jar probability $\theta$, the probability that the color $c$ on the current trial $t$ will be the same as the previous trial $t-1$ is computed as:

$$p(c_{t-1}=c_t)=\theta[(1-H)\theta+H(1-\theta)]+(1-\theta)[H\theta+(1-H)(1-\theta)]$$

The average switch probability is computed as

$$p(c_{t-1}\neq c_t)=1-p(c_{t-1}=c_t)$$

which we used to determine the correct hazard rate when jar probabilities = 1.

The optimal strategy in this task is to focus on the previous bead, and use your knowledge of the hazard rate to either predict the same color bead (in the low hazard condition), or predict the opposite color bead (in the high hazard condition). Any strategy that does not either always repeat or always switch does not maximize the possible amount of reward and would be indicative of probability matching-like strategies.

In this environment, we will examine whether people do indeed probability match (i.e., switch their predictions in proportion to the number of switches in the condition), whether performance is identical in both hazard rate conditions, whether people's information complexity correlates with predictive accuracy (see section 3.2), and how far people fall from the optimal bound as a function of their information complexity (see section 3.3). 

#### 2.2.2) Hidden Markov bead prediction task without jar strategy instructions

This second experiment is meant to test hypotheses 1-3 in section 1.2 above. A second set of 70 participants will perform a similar bead estimation task but with .01/.99 hazard rate conditions, and jar probabilities of 80% white/black beads. Again, subject will be explicitely told the hazard rate and jar probabilities at the start of each hazard rate block, and jars with 80% white/black beads will be displayed throughout the task. Similar to task above, the sequence of low probability events generating the jar switches and bead draws will be identical in the low and high hazard conditions, such that trials that mark jar switches/low probability bead colors appear in the low hazard rate condition will match the trials on which the jars repeat/low probability bead colors appear in the high hazard condition.

<img src="./figs/HiddenMarkovTask-14.png" alt="Drawing" style="width: 700px;"/>

In this environment, the best inference strategy is to use the history of past beads to infer which of the next jars is most likely to generate the next bead. As such, good strategies in this environment should encode information about more than just the last bead to infer the likely next jar.

This condition is closely matched to the task conditions in Filipowicz et al., 2020 and is meant to test whether people adjust their strategies appropriately, and use effective strategies in environments requiring inference over a latent variable (i.e., the jar).

#### 2.2.3) Hidden Markov bead prediction task with jar strategy instructions

This third experiment is meant to test hypothesis 4 in the section 1.2 above. A third group of 70 participants will perform an identical task as in 2.2.2, but with an added instruction informing them that the best strategy to maximize accuracy is to try and infer the jar that will produce the next bead, and not just the previous bead. This instruction is meant to mitigate the "one-back" strategy we observed in the high hazard rate condition in the pilot analyses of Filipowicz et al., 2020 data (see section 4). This instruction will be provided at the start of each hazard block along with the hazard rate and jar probabilities (which will still be .01/.99 and .8 respectively).

## 3) Planned analyses<a name='panalyses'/>


### 3.1) Empirical estimate of the information bottleneck

The problem of building effective mental models can be cast as a problem of "lossy compression": given certain encoding limitations, agents compress past information from their environment into "mental models" that they use to interact with the world. A substantial challenge when trying to measure the "optimality" of these mental models is that the functional form of this compression requires knowing an agent's mental model, something we do not have explicit access to. Furthermore, knowing how effectively this model is using its compressed information is also challenging as it requires knowing the optimal oracle for every given level of compression.

The information bottleneck framework offers solutions to both of these problems. This section will go over the information bottleneck method and how we will apply it to our planned experiments.

The goal of the information bottleneck method is to find the best tradeoffs between compression $\hat{X}$ of one random variable $X$ to predict another $Y$ given their joint distribution $p(X,Y)$. This compression can be obtained by solving the following equation:

$$L[p(\hat{X}|X)]=I(X;\hat{X})-\beta I(\hat{X};Y)$$

where $I(;)$ indicates the mutual information, $L[p(\hat{X}|X)]$ indicates the quatitity to be minimized, and $\beta$ is a Lagrange multiplier. Here, $I(X;\hat{X})$ corresponds to the amount of information our compression $\hat{X}$ retains about $X$, and $\beta I(\hat{X};Y)$ corresponds to the information we want to preserve about $Y$. The Lagrange multiplier $\beta$ controls the amount of information we want to preserve about $Y$. Solving this equation for a range of values of $\beta$, we obtain an information bottleneck curve that tells us the minimum compression we will need to obtain different levels of information about $Y$. When $\beta \rightarrow 0$, none of the past needs to be retained (since we do not need to retain any information about $Y$). Conversely, $\beta \rightarrow \infty$ implies that $\hat{X}$ ecodes the maximum it can from $X$ to predict $Y$ and the curve saturates at $I(\hat{X};Y)$.

Typically, the equations above are hard to solve analytically. However, a numerical solution, used in rate distortion theory, can be obtained using the Blahut-Arimoto algorithm. For a source $X$, this algorithm converges on an optimal compression $\hat{X}$ that minimizes the source's distortion $d(X,\hat{X})$ using the joint distribution $p(X,\hat{X})$. It is implemented by iterating between these two equations until convergence:

$$p_{t+1}(\hat{X})=\sum_{x \in X}{p_t(x)p_t(\hat{x}|x)}$$

$$p_{t+1}(\hat{X}|X)=\frac{p_t(\hat{x})\exp(-\beta d_t(x,\hat{X}))}{\sum_{\hat{x} \in \hat{X}}{p_t(\hat{x})\exp(-\beta d_t(x,\hat{x}))}}$$

where $\beta$ indicates the targeted slope in the rate distortion curve (i.e., the amount of distortion we are willing to accept in our signal).

In the context of the information bottleneck, $\beta$ has a similar function, indcating the amount of the variable $X$ we would need to retain a certain amount of information about $Y$. Tishby and colleagues demonstrate that the distortion $d(x,\hat{x})$ is equivalent to $D_{KL}[p(Y|X)\| p(Y|\hat{X})]$, where $D_{KL}$ is the KL divergence. Therefore, in order to obtain the minimum $p(\hat{X}|X)$ that maximizes $\beta I(\hat{X};Y)$ for a given value $\beta$, we can iterate through the following equations

$$p(\hat{X}|X) = \frac{p(\hat{X})}{Z(X,\beta)}\exp\Big[-\beta\sum_{y \in Y}{p(y|X)\log\frac{p(y|X)}{p(y|\hat{X})}}\Big]$$

where $Z(X,\beta)$ represents the normalizing factor $Z(X,\beta) = \sum_{\hat{x} \in \hat{X}}{p(\hat{x}) \exp \Big[\sum_{y \in Y}{p(y|x) \log \frac{p(y|x)}{p(y|\hat{x})}}\Big]}$,

$$p(Y|\hat{X}) = \frac{1}{p(\hat{x})} \sum_{x \in X}{p(y|x)p(\hat{x}|x)p(x)}$$

to ensure the markovian relation $\hat{X} \leftarrow X \leftarrow Y$ between past, compression and future, and 

$$p(\hat{X}) = \sum_{x \in X}{p(\hat{x}|x)p(x)}$$

Iterating through these three equations provides a distribution $p(\hat{X}|X)$ for a given value of $\beta$ that minimizes $I(X;\hat{X})$, while maximizing $\beta I(\hat{X};Y)$.

To estimate the information bottleneck for our data, we will use a python package we have written ([EMBO](https://pypi.org/project/embo/)) that empirically estimates the information bottleneck curve between two discrete sets of observations.

### 3.2) Applying the information bottleneck to human decision-making data

The information bottleneck method is trying to solve a very similar problem as humans in decision-making tasks: extract the most relevant information from the past $X_{past}$ to form some mental model $M$ that makes accurate predictions about the future $X_{future}$. The amount of information people encode from the past $I_{past}$ can be measured as the mutual information between past stimulus features and a participant's mental model:

$$I_{past} \equiv I(X_{past};M) \equiv \sum_{x}\sum_{m} p(x,m) \log_2\Big[\frac{p(x,m)}{p(x)p(m)}\Big]$$

Likewise, the accuracy of the mental model can be measured as the information the mental model carries about the task feature it is trying to predict:

$$I_{future} \equiv I(M; X_{future}) \equiv \sum_{x}\sum_{m} p(m,x) \log_2\Big[\frac{p(m,x)}{p(m)p(x)}\Big]$$

In order to measure these quantities we will need to define the parts of the task that correspond to subject "mental models" ($M$), "past" events ($X_{past}$), and "future" events ($X_{future}$).

#### 3.2.1) Proxy for "mental models"

We will use subject trial-by-trial responses $R$ as proxies for their mental models. The main assumptions with this method are that a) responses are veridical representations of subject mental models, and b) their mental models are stable throughout the period being measured.

#### 3.2.2) Defining "past" features

To assess the amount of information people are encoding from the task, we will measure $I_{past}$ using increasingly large windows of observations into the past. Technically this window could extend into the entire past, which would make our estimates of the mutual information unfeasible for both measuring human information complexity and empirically estimating the information bottleneck. However, there are a few properties of human memory limits and of the specific sequences we have contructed that make this more tractable.

First, research on tasks such as the "n-back" task show that, due to working memory limits, people are likely not encoding much more than ~3 trials into the past. This makes it such that we will be able to capture a substantial portion of the strategies people are likely to use with reasonably limited window sizes into the past.

Second, a property of the sequences we have constructed for our experiments is that the main information they contain can be extracted using a reasonably limited window into the past. This is especially true for the markov sequence with no latent variable, as all of the information needed to predict the future is contained in the previous trial. For this task environment, the information bottleneck only needs to be computed using the previous trial. Similarly, in the latent hidden markov task conditions, the hazard rate serves as a temporal filter, which limits the extent to which past observations are related to future observations. As such, increasing the window size into the past provides diminishing returns related to information about the future. Although an infinite window into the past is technically the most correct way to estimate the information bottleneck for a hidden markow environment, it will be well approximated by a more limited window size into the past.

To find the appropriate window to compute $I_{past}$ for both the empirial information bottleneck and subject information complexity, we will compute the mutual information between increasing windows of observations into the past and either future observations ($X_0$) for the information bottleneck, or subject predictions ($R_0$) for subject information complexity.

For the information bottleneck

$$I(X_{-n:-1}; X_0) \equiv \sum_{x_{-n:-1}}\sum_{x_0} p(x_{-n:-1},x_0) \log_2\Big[\frac{p(x_{-n:-1},x_0)}{p(x_{-n:-1})p(x_0)}\Big]$$

and for subject information complexity

$$I(X_{-n:-1}; R_0) \equiv \sum_{x_{-n:-1}}\sum_{r} p(x_{-n:-1},r_0) \log_2\Big[\frac{p(x_{-n:-1},r_0)}{p(x_{-n:-1})p(x_0)}\Big]$$

where $X_{-n:-1}$ corresponds to a window of $n$ previous trials, $X_0$ corresponds to the bead color on the current trial the subject is trying to predict, and $R_0$ correpsonds to the subjects current prediction, before observing $X_0$.

Because we have no specific priors as to where these values should fall, we will use the default uniform priors for all regression coefficients provided by the [brms](https://cran.r-project.org/web/packages/brms/index.html) package.

To estimate the information bottleneck in the markov condition, we will just compute the bottleneck considering one trial into the past. In both the hidden markov conditions, and to estimate subject information complexity, we will extend our window into the past as far as our estimates will allow before sampling issues make estimation unfeasible. This will be assed by examining how the mutual information estimates change as a function of window size. To correct for undersampling, the mutual information will be estimated using the [NSB method](https://papers.nips.cc/paper/1965-entropy-and-inference-revisited.pdf) (Nemenman, Shafee, & Bialek). Using this method, when undersampling becomes too problematic, the estimates drop dramatically. We will therefore compute mutual information values for window sizes up until we start seeing these undersampling issues, and use the window size with the highest estimate of the mutual information for our estimates of the empirical information bottleneck bound and subject information complexity.

#### 3.2.3) Defining "future" features

We define future features as the aspect of the task the subject is trying to predict. In the three tasks above, this corresponds to the color of the next bead. As such, a subject's $I_{future}$ will be computed as

$$I_{future} \equiv I(R_0; X_0) \equiv \sum_{x_0}\sum_{r_0} p(r_0,x_0) \log_2\Big[\frac{p(r_0,x_0)}{p(r_0)p(x_0)}\Big]$$

where $x_0$ indicates the color of the next bead, and $r_0$ indicates a person's prediction about the next bead.

#### 3.2.4) Approximations of mutual information

Mutual information can be heavily affected by the "curse of dimensionality", showing strong positive biases when data are undersampled. To correct for these biases, mutual information will be computed using the [NSB method](https://papers.nips.cc/paper/1965-entropy-and-inference-revisited.pdf) for entropy estimation.

Mutual information estimation error will be computed using bootstrap methods (see example in section 4).

### 3.3) Computing subject optimality

In the information bottleneck space, an optimal model is one in which its $I_{future}$ is on the information bottleneck bound for the model's given level of $I_{past}$. To measure how optimally each subject is trading off $I_{past}$ and $I_{future}$, we will measure the vertical distance between each subject's point in the information bottleneck space and the information bottleneck bound. We will use this "distance from the bound" metric to examine the extent to which information complexity relates to optimality.

### 3.4) Correlation measures

We plan at least two types of correlations to test our hypotheses: 

1. Correlations between $I_{past}$ and $I_{future}$ to assess the link between complexity and accuracy.

2. Correlations between $I_{past}$ and distance from the bound to measure links between complexity and optimality.

One issue is that both types of correlations depend on $I_{future}$, which has a range that increases as a function of $I_{past}$. This introduces the strong possibility of heteroskedasticity, where variance increases as a function of $I_{past}$. Indeed, this increase in variance is clearly present in our pilot analyses of the data from Filipowicz et al., 2020 (see section 4). 

Additionally, some of our hypothesis predict no correlation between $I_{past}$ and distance from the bound, meaning that we need a framework to measure evidence for the null.

To address both of these issues we will estimate correlations using Bayesian regression implemented in the [Stan](https://mc-stan.org/users/interfaces/rstan) language using the [brms](https://cran.r-project.org/web/packages/brms/index.html) package in R.

We will measure the posterior estimates for each $\beta$ below:

$$H_1: y=\beta_0+\beta_1I_{past}+N(0,\beta_{\sigma}I_{past})$$

where $y$ will be either $I_{future}$ or distance from the bound, $\beta_0$ corresponds to the intercept (which will measure the offset from the bound in our second correlation), $\beta_1$ measures the influence of $I_{past}$ on $y$, $N$ indicates gaussian noise, and $\beta_{\sigma}$ captures the heteroskedasticity.

This model will be compared to a simplified intercept only model, which we will use as our null:

$$H_0: y=\beta_0+N(0,\beta_{\sigma}I_{past})$$

To assess whether or not a correlation is present, we will compare the Bayes factor between these models:

$$BF_{10}=\frac{p(D|H_1)}{p(D|H_0)}$$

where $BF_{10}$ indicates the evidence for $H_1$ over $H_0$, and $D$ indicates the observed data. With this formulation, $BF_{10}$ > 1 will indicate evidence for the correlation model $H_1$ and $BF_{10}$ < 1 indicating evidence for the null model $H_0$. As suggested by [Lee et al., 2013](https://psycnet.apa.org/record/2014-14454-000), we will treat $BF_{10}$ values between 3-10 or below 1/3-1/10 as moderate evidence, and values > 10 or < .1 as strong evidence for $H_1$ and $H_0$ respectively.

## 4) Additional Information<a name='addInfo'/>

Below we include an example of the steps and functions we will use for our data analyses. These are what we have referred to as "pilot analyses" applied to data from Filipowicz et al., 2020.

### 4.1) Load data and show behavioral trends

The data come from 78 subjects who performed a task in which they were trying to predict which of two sources would generate a tone on the next trial. In this task, the "sources" correspond to the "jars" in the experiments outlined above, and the "tones" correspond to the "beads". Participants in this task were switched between different hazard rates, including low hazard rate trials (H = 0.01) and high hazard rate trials (H = 0.99). The probability that the tone would match the source was fixed at 0.80.

We note that there are four major difference between this task and our proposed tasks above:

1. Participants were instructed that low, intermediate, and high hazard rates would apply throughout the task, but were not explicitely told which hazard rate applied on any given trial.

2. There was an additional "intermediate" hazard rate of 0.3 in this task. We have omitted this condition from the analyses below given that we will not be focusing on this condition in our current experiments.

3. The hazard rates in this task were not segmented by blocks as they will be in the tasks above: the hazard rates switched at unannounced points throughout the task.

4. The sequence of tones the subjects heard was identical across subjects. However, the sequences of low and high hazard tones were not matched as they will be in our experiments above.

### 4.2) Determine appropriate window size to compute information bottleneck

Next we compute the mutual information between past observations (`X` below) and the source (`Z`), since this was what subjects were trying to predict.

```python
# Test bounds in each session
def get_windowed_mi(x,z,nw):
    '''
    Function that windows of observations from a sequence x offset from y and computes the MI
    Input:
        - x: sequence of "past" observations to be windowed
        - z: sequence that serves as the "future" events
        - nw: maximum window size
    Output:
        - mis: NSB estimated mutual informtion values for each window size
    '''
    mis = np.zeros(nw+1) #vector of mutual informations for different window sizes
    for w in np.arange(1,nw+1):
        xw,zw = ut.get_windowed_x(x,z,w=w)         # use convenience function from utilities to window x
        mis[w] = ut.mutual_inf_nsb(xw,zw,[2**w,2]) # use convenience function from utilities to compute NSB mutual info
    return(mis)

# Get all trials in all three sessions that match low and high hazard rates
nw = 10 #number of windows into the past we want to compute

#Since everyone saw the same sequence, we can speed this up by subselecting one example of all the trials the first subject saw
adat_trials = adat[0:3000]

# Computed windowed predictive information in the low hazard rate
adat_mis_h1 = get_windowed_mi(adat_trials['X'].loc[adat_trials['H'] == .01]-1,
                              adat_trials['Z'].loc[adat_trials['H'] == .01],nw)

# Computed windowed predictive information in the low hazard rate
adat_mis_h9 = get_windowed_mi(adat_trials['X'].loc[adat_trials['H'] == .99]-1,
                              adat_trials['Z'].loc[adat_trials['H'] == .99],nw)

# Plot predictive information as a function of window size
plt.figure(1,(8,6))
plt.plot(np.arange(nw+1),adat_mis_h1,'-o',label='H = 0.01')
plt.plot(np.arange(nw+1),adat_mis_h9,'-o', label = "H = 0.99")
plt.ylabel('Mutual Information between\n$X_{past}$ and $X_{future}$',size=15)
plt.xlabel('Window size of $X$ into the past',size=15)
plt.legend()
```

### 4.3) Compute the empirical information bottleneck

We can see that the curves peak around window size 8. We will use this size for our empirical estimate of the information bottleneck. We will also compute a bound with only one trial back to show where subjects would fall if they were prone to using a "one-back" type strategy.

```python
%%time
def get_windowed_bound(x,z,w,mb=50,nb=2000,p=8):
    '''
    Function that computes the information bottleneck bound between two discrete variables
    Input:
        - x: sequence of past events
        - z: sequence of future events
        - w: size of the window to be applied to x
        - mb: maximum value of the lagrange multiplier beta
        - nb: number of values of beta between 0 and mb to be run through the Blahaut-Arimoto algorithm
        - p: number of cores to use in the computation (to help with speed)
    Output:
        - 4 element tuple with a sequence of ipast and ifuture values corresponding to the information bottleneck for full and one back bounds
    '''
    # Get desired windowed x and corresponding z
    xw,zw = ut.get_windowed_x(x,z,w=w)
    
    # Compute the information bottleneck using EMBO for the window size w specified above
    ipw,ifw,betasw = EmpiricalBottleneck(xw,zw,minsize=True,processes=p,maxbeta=mb,numbeta=nb).get_empirical_bottleneck() 
    
    # Compute the information bottleneck using EMBO for the window size w specified above
    x1b,z1b = ut.get_windowed_x(x,z,w=1)
    ip1b,if1b,betas1b = EmpiricalBottleneck(x1b,z1b,minsize=True,processes=p,maxbeta=mb,numbeta=nb).get_empirical_bottleneck() 
   
    # Return informtion bottleneck
    return((ipw,ifw,ip1b,if1b))

# For speed, see if bottleneck has already been computed
try:
    # Low hazard full and 1-back bounds
    ipw_h1,ifw_h1,ipw_h1_1back,ifw_h1_1back = pickle.load(open('ipw_h1_bounds.pkl','rb'))
    
    # High hazard full and 1-back bounds
    ipw_h9,ifw_h9, ipw_h9_1back,ifw_h9_1back = pickle.load(open('ipw_h9_bounds.pkl','rb'))
    
except:
    # If no objects exist, compute full and one-back bounds for each hazard rate
    #Low hazard full and 1-back bounds
    ipw_h1,ifw_h1,ipw_h1_1back,ifw_h1_1back = get_windowed_bound(adat_trials['X'].loc[adat_trials['H'] == .01]-1,
                                                                 adat_trials['Z'].loc[adat_trials['H'] == .01],8)
    pickle.dump((ipw_h1,ifw_h1,ipw_h1_1back,ifw_h1_1back),open('ipw_h1_bounds.pkl','wb')) 
    
    #High hazard full and 1-back bounds
    ipw_h9,ifw_h9,ipw_h9_1back,ifw_h9_1back = get_windowed_bound(adat_trials['X'].loc[adat_trials['H'] == .99]-1,
                                                                 adat_trials['Z'].loc[adat_trials['H'] == .99],8)
    pickle.dump((ipw_h9,ifw_h9,ipw_h9_1back,ifw_h9_1back),open('ipw_h9_bounds.pkl','wb')) 
    
# Plot bounds for each hazard rate
plt.figure(1,(15,5))

# Low hazard rate
plt.subplot(1,2,1)
plt.plot(ipw_h1,ifw_h1,'-',color=cmap(0),label='full bound')
plt.plot(ipw_h1_1back,ifw_h1_1back,'--',color=cmap(0), label='one-back bound')
plt.ylim([0,.8])
plt.xlabel('$I_{past}$',size=15)
plt.ylabel('$I_{future}$',size=15)
plt.title('Low hazard rate',size=20)
plt.legend()

# High hazard rate
plt.subplot(1,2,2)
plt.plot(ipw_h9,ifw_h9,'-',color=cmap(1),label='full bound')
plt.plot(ipw_h9_1back,ifw_h9_1back,'--',color=cmap(1), label='one-back bound')
plt.ylim([0,.8])
plt.xlabel('$I_{past}$',size=15)
plt.ylabel('$I_{future}$',size=15)
plt.title('High hazard rate',size=20)
plt.legend()
```

### 4.3) Compute subject information complexity ($I_{past}$) and predictive accuracy ($I_{future}$)

Using a similar method to find the right window size for our information bottleneck bound, we will compute the information participants encode from increasing windows of observationas into the past. 

```python
%%time
def get_windowed_mi_subs(sdat_all,nw):
    '''
    Function to compute windowed mutual information values for all subjects
    Input:
        - sdat_all: data frame with all of the subject tones, responses, and sources
        - nw: maximum window size to be considered
    Output:
        - Ipast for each window size and Ifuture
    '''
    # Create dictionaries to keep track of variables with subject IDs as the keys
    subs = pd.unique(sdat_all['Subject'])
    ip_sub = {}
    if_sub = {}
    # Loop through each subject and get their Ipast for windows up to nw and their ifuture
    for subi,sub in enumerate(subs):
        sdat = sdat_all[sdat_all['Subject'] == sub]                   # Get data from specific subject 
        ip_sub[sub] = get_windowed_mi(sdat['X']-1,sdat['Resp'],nw)    # compute Ipast using function defined above 
        if_sub[sub] = ut.mutual_inf_nsb(sdat['Resp'],sdat['Z'],[2,2]) # compute Ifuture
    # Return dictionaries of ipast and ifuture values
    return(ip_sub,if_sub)

try:
    # Check if subject mutual information values have already been computed
    ips_sub_h1,if_sub_h1 = pickle.load(open('ipf_sub_h1.pkl','rb')) #load windowed mutual information values for H=.01
    ips_sub_h9,if_sub_h9 = pickle.load(open('ipf_sub_h9.pkl','rb')) #load windowed mutual information values for H=.99
except:
    # If not, compute them below
    # Low hazard rate
    ips_sub_h1,if_sub_h1 = get_windowed_mi_subs(adat_h19.loc[adat['H'] == .01],10)
    pickle.dump((ips_sub_h1,if_sub_h1),open('ipf_sub_h1.pkl','wb'))
    
    # High hazard rate
    ips_sub_h9,if_sub_h9 = get_windowed_mi_subs(adat_h19.loc[adat['H'] == .99],10)
    pickle.dump((ips_sub_h9,if_sub_h9),open('ipf_sub_h9.pkl','wb'))
    
# Plot ipast as a function of window size for each subject (different colors)
plt.figure(1,(15,5))
# Low hazard rate
plt.subplot(1,2,1)
for sub in pd.unique([key for key in ips_sub_h1.keys()]):
    plt.plot(np.arange(11),ips_sub_h1[sub],'-o')
    plt.ylabel('Subject $I_{past}$',size=15)
    plt.xlabel('Window size of $X$ into the past',size=15)
    plt.title('Low hazard rate',size=20)
    plt.ylim([0,1])

# high hazard rate
plt.subplot(1,2,2)
for sub in pd.unique([key for key in ips_sub_h9.keys()]):
    plt.plot(np.arange(11),ips_sub_h9[sub],'-o')
    plt.ylabel('Subject $I_{past}$',size=15)
    plt.xlabel('Window size of $X$ into the past',size=15)
    plt.title('High hazard rate',size=20)
    plt.ylim([0,1])
```

Using these values, we can compute a subject's $I_{past}$ by identifying the window size with the maximum mutual information value. To compute this quantity more robustly, and capture the error range in our estimates, we can use a bootstrap method to repeat this process and provide a distribution of maximum mutual information values for each subject. We can then take the mean of this distribution for each subject ot get their bootstrap-estimated $I_{past}$, and use the standard deviation to compute a 95% confidence interval.

```python
%%time
def get_bootstrapped_samples(dat,wips,nboot):
    '''
    Function to compute and return bootstrapped distributions of Ipast and Ifuture values for each subject
    Input:
        - dat: all subject data
        - wips: matrix of mutual information by widow size values for each subject
        - nboot: number of requested bootstrap iterations
    Output:
        - dictionaries of bootstrapped distributions of Ipast and Ifuture values
    '''
    # Get subject IDs and initialize dictionaries
    subs = pd.unique(dat['Subject'])
    ip_boot_mi = {}
    if_boot_mi = {}
    
    # Loop through each subject and get bootstrapped estimates
    for subi,sub in enumerate(subs):
        sdat = dat[dat['Subject'] == sub]   # Get subject data
        w = wips[sub].argmax()              # Get subject's maximum window size
        x = np.array(sdat['X']-1)           # Get tones, responses, and sources for the subject
        r = np.array(sdat['Resp'])
        z = np.array(sdat['Z'])
        xw,rw = ut.get_windowed_x(x,r,w=(w+1)) # Window the tones and responses appropriately
        ip_boot_mi[sub] = np.zeros(nboot)      # Initialize arrays of mutual information values
        if_boot_mi[sub] = np.zeros(nboot)
        
        # Run bootstrap procedure
        for boot in np.arange(nboot):
            idx = np.random.choice(np.arange(len(xw)),size=len(xw),replace=True)    # Select random indicies with replacement
            ip_boot_mi[sub][boot] = ut.mutual_inf_nsb(xw[idx],rw[idx],[2**(w+1),2]) # Use indexed data to compute Ipast
            if_boot_mi[sub][boot] = ut.mutual_inf_nsb(r[idx],z[idx],[2,2])          # Use indexed data to compute Ifuture
    # Return dictionaries of Ipast and Ifuture distributions
    return(ip_boot_mi,if_boot_mi)

try:
    # Check to see if bootstrapped samples have already been computed
    ip_h1_boot, if_h1_boot = pickle.load(open('ip_if_h1_boot.pkl','rb')) # Low hazard
    ip_h9_boot, if_h9_boot = pickle.load(open('ip_if_h9_boot.pkl','rb')) # High hazard
except:
    # If not, compute and save below
    nboot = 1000 #number of bootstrapped samples
    #Low Hazard
    ip_h1_boot, if_h1_boot = get_bootstrapped_samples(adat_h19.loc[adat['H'] == .01],ips_sub_h1,nboot)
    pickle.dump((ip_h1_boot, if_h1_boot),open('ip_if_h1_boot.pkl','wb'))
    
    # High hazard
    ip_h9_boot, if_h9_boot = get_bootstrapped_samples(adat_h19.loc[adat['H'] == .99],ips_sub_h9,nboot)
    pickle.dump((ip_h9_boot, if_h9_boot),open('ip_if_h9_boot.pkl','wb'))
```

Points indicate each subjects estimated $I_{past}$ and $I_{future}$ and the lines indicate 95% bootstrapped confidence intervals. The statistics reported are $rho$ and $p$ values from Spearman correlations.

```python
# Get mean Ipast/Ifuture values per subject
# Low hazard Ipast values and 95% confidence intervals
ip_mean_h1 = np.array([np.mean(ip_h1_boot[key]) for key in ip_h1_boot.keys()])
ip_95CI_h1 = np.array([np.std(ip_h1_boot[key])*1.96 for key in ip_h1_boot.keys()])

# Low hazard Ifuture values and 95% confidence intervals
if_mean_h1 = np.array([np.mean(if_h1_boot[key]) for key in if_h1_boot.keys()])
if_95CI_h1 = np.array([np.std(if_h1_boot[key])*1.96 for key in if_h1_boot.keys()])

# High hazard Ipast values and 95% confidence intervals
ip_mean_h9 = np.array([np.mean(ip_h9_boot[key]) for key in ip_h9_boot.keys()])
ip_95CI_h9 = np.array([np.std(ip_h9_boot[key])*1.96 for key in ip_h9_boot.keys()])

# High hazard Ifuture values and 95% confidence intervals
if_mean_h9 = np.array([np.mean(if_h9_boot[key]) for key in if_h9_boot.keys()])
if_95CI_h9 = np.array([np.std(if_h9_boot[key])*1.96 for key in if_h9_boot.keys()])

# Plot Ipast vs Ifuture in both hazard rate conditions
plt.figure(1,(15,5))
# Low hazard
plt.subplot(1,2,1)
r,p=spearmanr(ip_mean_h1,if_mean_h1)
plt.errorbar(ip_mean_h1,if_mean_h1,xerr=ip_95CI_h1,yerr = if_95CI_h1,
             marker='o',markersize=9,mec='white',color=cmap(0),ls='none')
plt.ylabel('Subject Predictive Accuracy\n$I_{future}$',size=15)
plt.xlabel('Subject Information Complexity\n$I_{past}$',size=15)
plt.title('Low Hazard Rate\nrho={},p={}'.format(np.round(r,2),p),size=20)
    
# High hazard
plt.subplot(1,2,2)
r,p=spearmanr(ip_mean_h9,if_mean_h9)
plt.errorbar(ip_mean_h9,if_mean_h9,xerr=ip_95CI_h9,yerr = if_95CI_h9,
             marker='o',markersize=9,mec='white',color=cmap(1),ls='none')
plt.xlabel('Subject Information Complexity\n$I_{past}$',size=15)
plt.title('High Hazard Rate\nrho={},p={}'.format(np.round(r,2),p),size=20)
```

### 4.5) Measure subject optimality as distance from the information bottleneck bound

To measure how optimally people use the information they encode, we can look at how their $I_{past}$ and $I_{future}$ compare to the optimal information bottleneck bound. We also plot the "one-back" bound to see if people who fall off the full bound adopt "one-back"-like strategies.

```python
plt.figure(1,(15,5))
# Low hazard
plt.subplot(1,2,1)
r,p=spearmanr(ip_mean_h1,if_mean_h1)
plt.plot(ipw_h1,ifw_h1,'-k',label='full bound')
plt.plot(ipw_h1_1back,ifw_h1_1back,'--k',label='one-back bound')
plt.errorbar(ip_mean_h1,if_mean_h1,xerr=ip_95CI_h1,yerr = if_95CI_h1,
             marker='o',markersize=9,mec='white',color=cmap(0),ls='none')
plt.ylabel('Subject Predictive Accuracy\n$I_{future}$',size=15)
plt.xlabel('Subject Information Complexity\n$I_{past}$',size=15)
plt.title('Low Hazard Rate',size=20)
plt.ylim([0,.8])
    
# High hazard
plt.subplot(1,2,2)
r,p=spearmanr(ip_mean_h9,if_mean_h9)
plt.plot(ipw_h9,ifw_h9,'-k',label='full bound')
plt.plot(ipw_h9_1back,ifw_h9_1back,'--k',label='one-back bound')
plt.errorbar(ip_mean_h9,if_mean_h9,xerr=ip_95CI_h9,yerr = if_95CI_h9,
             marker='o',markersize=9,mec='white',color=cmap(1),ls='none')
plt.xlabel('Subject Information Complexity\n$I_{past}$',size=15)
plt.title('High Hazard Rate',size=20)
plt.legend()
plt.ylim([0,.8])
```

From these plots, it looks as though subjects in the low-hazard rate condition vary in their accuracy, but fall an equal distance from the bound regardless of their information complexity. However, subjects in the high hazard condition seem to fall farther from the optimal bound as they get more complex, following the one-back bound more closely.

We can measure this "dstance from the bound" as the vertical distance between each subject's point in information bottleneck space and the full bound.

```python
def deltaBound(ib_ipast,ib_ifuture,p_ipast,p_ifuture):
    ''' 
    Function to calculate vertical distance from the bound between an empirical IB and participant predictive info
    ib_ipast: ipast of empirical IB (x of convex hull)
    ib_ifuture: ifuture of empirical IB (y of convex hull)
    p_ipast: participant ipast (uncorrected)
    p_ifuture: participant ifuture (uncorrected)
    
    Returns participant ifuture minus the empirical bound (more negative = farther away from the bound)
    '''
    ind = np.argwhere(np.array(ib_ipast) > p_ipast)[0][0]
    slp = (ib_ifuture[ind]-ib_ifuture[ind-1])/(ib_ipast[ind]-ib_ipast[ind-1])
    intercept = ib_ifuture[ind]-(slp*ib_ipast[ind])
    #Return distance between participant Ifuture and interpolated bound
    return p_ifuture - ((p_ipast*slp)+intercept)

# Compute distance from the bound for each subject in each hazard rate condition
dbs1 = np.zeros(len(ip_mean_h1))
dbs9 = np.zeros(len(ip_mean_h9))
for i in np.arange(len(ip_mean_h1)):
    dbs1[i] = deltaBound(ipw_h1,ifw_h1,ip_mean_h1[i],if_mean_h1[i])
    dbs9[i] = deltaBound(ipw_h9,ifw_h9,ip_mean_h9[i],if_mean_h9[i])

# Plot distance from the bound for each hazard rate
plt.figure(1,(15,5))
# Low hazard
plt.subplot(1,2,1)
r,p=spearmanr(ip_mean_h1,dbs1)
plt.plot([0,1],[0,0],'-k',label='full bound')
plt.errorbar(ip_mean_h1,dbs1,xerr=ip_95CI_h1,yerr = if_95CI_h1,
             marker='o',markersize=9,mec='white',color=cmap(0),ls='none')
plt.ylabel('Subject Distance from Full Bound',size=15)
plt.xlabel('Subject Information Complexity\n$I_{past}$',size=15)
plt.title('Low Hazard Rate\nrho={},p={}'.format(np.round(r,2),np.round(p,3)),size=20)
plt.xlim([0,1])
plt.ylim([-.5,.12])
    
# High hazard
plt.subplot(1,2,2)
r,p=spearmanr(ip_mean_h9,dbs9)
plt.plot([0,1],[0,0],'-k',label='full bound')
plt.errorbar(ip_mean_h9,dbs9,xerr=ip_95CI_h9,yerr = if_95CI_h9,
             marker='o',markersize=9,mec='white',color=cmap(1),ls='none')
plt.xlabel('Subject Information Complexity\n$I_{past}$',size=15)
plt.title('High Hazard Rate\nrho={},p={}'.format(np.round(r,2),p),size=20)
plt.xlim([0,.8])
plt.ylim([-.5,.12])
plt.legend()
```
